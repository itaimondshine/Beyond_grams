```python
import spacy
import pandas as pd

# Load the spaCy English model
language = "arabic"
nlp = spacy.load('es_core_news_sm')
```

```python
def lemmatize_text(text: str) -> str:
    doc = nlp(text)
    lemmatized_tokens = [token.lemma_ for token in doc]
    lemmatized_text = ' '.join(lemmatized_tokens)
    return lemmatized_text
```

```python

path = f"/Users/itaimondshine/PycharmProjects/NLP/eval_metrics/app/xlsum/{language}/pred.csv"
df = pd.read_csv(path)
df['label_lemma'] = df['label'].apply(lambda f: lemmatize_text(f))
df['gpt_prediction_lemma'] = df['gpt_prediction'].apply(lambda f: lemmatize_text(f))
df['gemini_prediction_lemma'] = df['gemini_prediction'].apply(lambda f: lemmatize_text(f))
df.to_csv(path, index=False)
```

```python
import stanza

stanza.download('zh', processors='lemma, tokenize')
nlp = stanza.Pipeline(lang='zh', processors='lemma, tokenize')

import stanza
def lemmaatize(text):
    doc = nlp(text)
    lemmatize = ''.join([word.lemma for sent in doc.sentences for word in sent.words])
    return lemmatize
```
