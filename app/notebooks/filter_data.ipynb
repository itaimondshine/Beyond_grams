```python
LANGUAGE = 'spanish'
```

```python
import pandas as pd

new_data = pd.read_csv(f"/Users/itaimondshine/PycharmProjects/NLP/eval_metrics/app/app_data/data/{LANGUAGE}/pred.csv")
old_data = pd.read_csv(f"/Users/itaimondshine/PycharmProjects/NLP/eval_metrics/app/old_data/filtered/{LANGUAGE}.csv")

new_data = new_data[~new_data['q_idx'].isin(old_data['q_idx'])]
all_data = pd.concat([new_data, old_data], axis=0)
all_data = all_data.drop_duplicates(subset=['q_idx'])
# indonesian = pd.read_csv("/Users/itaimondshine/PycharmProjects/NLP/eval_metrics/app/app_data/cleaned/indonesian.csv")
```

```python
all_data
```

```python
new_rows = []
q_idexes= []
for _, row in all_data.iterrows():
    q_idx = row['q_idx']
    label = row['label'].split(".")
    gpt_prediction = row['gpt_prediction'].split(".")
    gemini_prediction = row['gemini_prediction'].split(".")
    
    print(gemini_prediction)
    print(gpt_prediction)
    print(label)
    if len(row['gemini_prediction']) < 600 and len(row['gpt_prediction']) < 600 and len(label) < 7:
        print(label)
        q_idexes.append(q_idx)
        new_rows.append({'q_idx': q_idx, 'label': row['label'], 'gpt_prediction': row['gpt_prediction'], 'gemini_prediction': row['gemini_prediction']})
```

```python
print(len(new_rows))
```

```python
pd.DataFrame(new_rows).to_csv(f"/Users/itaimondshine/PycharmProjects/NLP/eval_metrics/app/app_data/cleaned/{LANGUAGE}.csv", index=False)
```
