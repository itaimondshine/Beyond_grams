```python
from collections import Counter

def normalize_coherence(documents, field="consistence_gemini"):
    # Group by n_sample

    grouped = {}
    for doc in documents:
        grouped.setdefault(doc["n_sample"], []).append(doc)
    
    print(len(grouped))
    # Process each group
    for n_sample, docs in grouped.items():
        values = [doc[field] for doc in docs]
        counter = Counter(values)
        
        # If there are exactly 3 unique values
        if len(counter) == 2:
            # Find the most common value
            common_value, common_count = counter.most_common(1)[0]
            
            # If two values are the same, update the third one
            if common_count == str(2) or common_count == 2:
                for doc in docs:
                    if doc[field] != str(common_value):
                        print("here2")
                        doc[field] = str(common_value)   # Normalize
    
    return documents
```

```python
import pandas as pd

def adjust_grades(df, grade_columns):
    """
    Adjusts grades in specified columns by:
    1. Replacing both values with -1 if there are two grades that differ by more than 1.
    2. Replacing an outlier with -1 if there are three grades and one is far (≥2 away from the closest).
    """
    def process_group(group):
        for col in grade_columns:
            values = group[col].astype(int)  # Convert to integers
            unique_values = sorted(values.unique())  # Get unique sorted values
            
            if len(unique_values) == 2:
                # Case 1: Two grades, difference > 1
                if abs(unique_values[0] - unique_values[1]) > 1:
                    group[col] = -1
            
            elif len(unique_values) == 3:
                # Case 2: Three grades, one is an outlier
                v1, v2, v3 = unique_values  # Sorted values
                if abs(v1 - v2) <= 1 and abs(v2 - v3) > 1:  
                    # v3 is the outlier → Replace it
                    group.loc[group[col] == v3, col] = -1
                elif abs(v2 - v3) <= 1 and abs(v1 - v2) > 1:  
                    # v1 is the outlier → Replace it
                    group.loc[group[col] == v1, col] = -1
                elif abs(v2-v1) == 1 and abs(v3-v2) == 1 and abs(v3-v1) ==2:
                     group.loc[group[col] == v1, col] = -1
                     group.loc[group[col] == v2, col] = -1
                     group.loc[group[col] == v3, col] = -1

        return group

    return df.groupby('n_sample', group_keys=False).apply(process_group)

# Convert documents to DataFrame
df = pd.DataFrame(documents)

# Define grade columns
grade_columns = ['coherence_gpt', 'coherence_gemini', 'consistence_gpt', 'consistence_gemini']

# Ensure grade columns are numeric
df[grade_columns] = df[grade_columns].apply(pd.to_numeric, errors='coerce')

# Apply the adjustment function
df = adjust_grades(df, grade_columns)

# Convert back to list of dictionaries
updated_documents = df.to_dict(orient='records')

# Print or use updated_documents
print(updated_documents)
```

```python

def replace_none_with_avg(lst):
    """Replace None values in a list of three elements with the average of the others."""
    if lst.count(None) == 3:
        return [0, 0, 0]  # Handle edge case where all values are None
    valid_numbers = [int(x) for x in lst if x is not None]
    avg = np.mean(valid_numbers) if valid_numbers else 0
    return [x if x is not None else avg for x in lst]

# Example usage
import pandas as pd

# df = pd.DataFrame({'stacked': [[None, 2, 4], [1, None, 3], [None, None, 5], [None, None, None]]})
melted_df['stacked'] = melted_df['stacked'].apply(replace_none_with_avg)
```

```python
def has_identical_numbers(lst):
    if lst is None:  # Handle None values if they exist
        return False
    # Remove None values
    lst = [x for x in lst if x is not None]
    # Check for duplicates
    number_counts = {n: lst.count(n) for n in lst}
    # Return True if any number has a count of 2 or more
    return any(count > 1 for count in number_counts.values())

# Apply the function to filter the DataFrame
df_filtered = melted_df[melted_df['stacked'].apply(has_identical_numbers)]
```

```python
from rouge_score import rouge_scorer
```

```python
def rouge_score(prediction, label):
    scorer = rouge_scorer.RougeScorer(
        ["rouge1", "rouge2", "rouge3", "rougeL"], use_stemmer=True
    )
    return compute_rouge(scorer, prediction, label)
```

```python
def compute_rouge(scorer, prediction, label):
    score = scorer.score(prediction, label)
    return (
        score["rouge1"].fmeasure,
        score["rouge2"].fmeasure,
        score["rouge3"].fmeasure,
        score["rougeL"].fmeasure,
    )
```
